## Visualización de Datos con Matplotlib
- Para visualizar datos en NumPy, una de las bibliotecas más utilizadas es Matplotlib. A continuación, te muestro cómo puedes crear gráficos básicos a partir de arrays de NumPy.


```python
import numpy as np
import matplotlib.pyplot as plt

# Paso 1: Crear arrays con datos de ventas mensuales
meses = np.array(['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio',
                  'Julio', 'Agosto', 'Septiembre', 'Octubre', 'Noviembre', 'Diciembre'])
ventas_A = np.array([150, 200, 250, 300, 220, 210, 180, 190, 230, 240, 280, 300])
ventas_B = np.array([180, 210, 230, 250, 270, 260, 240, 250, 270, 290, 310, 330])
ventas_C = np.array([200, 220, 240, 260, 280, 300, 320, 340, 360, 380, 400, 420])

# Paso 2: Reshape y agregación
# Agregar las ventas de los tres productos en un array 2D
ventas_totales = np.array([ventas_A, ventas_B, ventas_C])
print("Ventas Totales (Array 2D):")
print(ventas_totales)

# Calcular la suma de ventas por mes
suma_ventas_por_mes = ventas_totales.sum(axis=0)
print("\nSuma de Ventas por Mes:")
print(suma_ventas_por_mes)

# Calcular la media de ventas por producto
media_ventas_por_producto = ventas_totales.mean(axis=1)
print("\nMedia de Ventas por Producto:")
print(media_ventas_por_producto)

# Paso 3: Transposición y operaciones avanzadas
# Transponer el array de ventas para obtener ventas mensuales por productos
ventas_transpuesta = ventas_totales.T
print("\nVentas Transpuesta (Ventas Mensuales por Productos):")
print(ventas_transpuesta)

# Calcular el producto punto entre ventas_A y ventas_B
producto_punto_AB = np.dot(ventas_A, ventas_B)
print("\nProducto Punto entre Ventas A y B:")
print(producto_punto_AB)

# Paso 4: Análisis de ventas
# Encontrar el mes con las ventas máximas para cada producto
mes_max_ventas_A = meses[np.argmax(ventas_A)]
mes_max_ventas_B = meses[np.argmax(ventas_B)]
mes_max_ventas_C = meses[np.argmax(ventas_C)]
print("\nMes con las Ventas Máximas:")
print(f"Producto A: {mes_max_ventas_A}")
print(f"Producto B: {mes_max_ventas_B}")
print(f"Producto C: {mes_max_ventas_C}")

# Encontrar el total de ventas del año para cada producto
total_ventas_A = ventas_A.sum()
total_ventas_B = ventas_B.sum()
total_ventas_C = ventas_C.sum()
print("\nTotal de Ventas del Año:")
print(f"Producto A: {total_ventas_A}")
print(f"Producto B: {total_ventas_B}")
print(f"Producto C: {total_ventas_C}")

# Paso 5: Elementos únicos y sus conteos (Ventas_A)
unique_elements_A, counts_A = np.unique(ventas_A, return_counts=True)
print("\nElementos Únicos en Ventas A y sus Conteos:")
print("Elementos Únicos:", unique_elements_A)
print("Conteos:", counts_A)

# Paso 6: Vistas y Copias
# Crear una vista de ventas_A
vista_ventas_A = ventas_A[:5]
print("\nVista de las primeras 5 ventas de A:", vista_ventas_A)

# Crear una copia de ventas_A
copia_ventas_A = ventas_A[:5].copy()
print("Copia de las primeras 5 ventas de A:", copia_ventas_A)

# Modificar la vista y la copia
vista_ventas_A[0] = 999
copia_ventas_A[0] = 555

print("Ventas A después de modificar la vista:", ventas_A)
print("Vista modificada:", vista_ventas_A)
print("Copia modificada:", copia_ventas_A)

# Paso 7: Broadcasting y Operaciones Lógicas
# Aplicar una operación de broadcasting para incrementar las ventas en un 10%
incremento = 1.10
ventas_A_incrementadas = ventas_A * incremento
print("\nVentas A incrementadas en un 10%:")
print(ventas_A_incrementadas)

# Filtrar ventas mayores a 250
ventas_mayores_250 = ventas_A[ventas_A > 250]
print("\nVentas mayores a 250:")
print(ventas_mayores_250)

# Paso 8: Selección de subconjuntos de elementos
# Seleccionar las ventas de los primeros 6 meses de cada producto
ventas_primer_semestre = ventas_totales[:, :6]
print("\nVentas del primer semestre de cada producto:")
print(ventas_primer_semestre)

# Paso 9: Visualización
# Gráfica de ventas totales por mes
plt.figure(figsize=(10, 5))
plt.plot(meses, suma_ventas_por_mes, marker='o', linestyle='-', color='b', label='Ventas Totales')
plt.xlabel('Meses')
plt.ylabel('Ventas Totales')
plt.title('Ventas Totales por Mes')
plt.legend()
plt.grid(True)
plt.show()

# Gráfica de ventas por producto
plt.figure(figsize=(10, 5))
plt.plot(meses, ventas_A, marker='o', linestyle='-', color='r', label='Producto A')
plt.plot(meses, ventas_B, marker='o', linestyle='-', color='g', label='Producto B')
plt.plot(meses, ventas_C, marker='o', linestyle='-', color='b', label='Producto C')
plt.xlabel('Meses')
plt.ylabel('Ventas')
plt.title('Ventas por Producto y Mes')
plt.legend()
plt.grid(True)
plt.show()
```
